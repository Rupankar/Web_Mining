```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.request import urlparse
import pandas as pd
from urllib.error import HTTPError
from urllib.error import URLError 
import pandas as pd
```

```python
#starting url
input_url = "https://www.google.com"
#enter number
num=3

#get url domain
current_url_domain = urlparse(input_url).netloc

#current domain
current_doamin = current_url_domain.split(".")

#list to store the new links,headings and parent_urls
links = []
headings = []
parent_urls = []

try:

  def crawler(input_site):
    new_url = set()

    html=urlopen(input_site)  
    bs=BeautifulSoup(html.read(),'html.parser')
    heading = bs.title.get_text()                                                 #get heading of the webpage

    for link in bs.find_all('a'):                                                 #access all anchor tags
      current_link = link.get('href')      

      if(current_link != "" and current_link!=None and "http" in current_link):   #check that the link is not null or empty or relative link
        if(current_doamin[1] not in current_link and current_link not in links):  #check that the link do not belong to the same domain as the initial webpage and not already traversed
          parent_url = urlparse(current_link).netloc                              #get parent url of the current visited link
          links.append(current_link)                                              #add link to the list
          print(current_link)                                                     #print the link
          new_url.add(current_link)                                               
          headings.append(heading)                                                #add the heading to the list                                                
          parent_urls.append(parent_url)                                          #add the parent url to the list

    return new_url

  new_links = []                                                                  #visit the websites in BFS logic
  new_links.append(input_url)
  for i in range(num):
    for count in range(len(new_links)):
      url = new_links.pop(0)
      urls = crawler(url)
      for j in urls:
        new_links.append(j)

except HTTPError as e:
  print ('HTTP Error',e)
except URLError as e:
  print('Server could not be found')
```

```python
#create df of the link, heading and parent url

links_df = pd.DataFrame(
    {'links': links,
     'headings': headings,
     'parent_url': parent_urls
    })
```
